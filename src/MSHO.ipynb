```python
import os
import time
save_dir = f'../test/MSHO-new-jSO'
for prob_idx in range(23, 31):
    try:
        time_str = time.strftime("%Y%m%d%H%M%S", time.localtime())
        # create a folder to save result
        # save_dir = f'../test/{time_str}'
        
        if not os.path.exists(save_dir):
            os.makedirs(save_dir)
        
        save_result_dir = f'{save_dir}/f{prob_idx}.dat'
        
        for _ in range(3):
            # Initialize the algorithm
            algo = MSHO(name="IHBKaiser", prob=CEC2017(i=prob_idx, dim=30), new_LS=True,gen_length=30, MAX_FES=500000, BASE_POPSZ=100, learning=True, dynamic_pop=True, phasethree=False)
            
            # Run the algorithm
            res, acep = algo.run()
            
            # Check if result is valid
            print(f'Result: {res}, Acceptance: {acep}')  # Debug output
            
            # Write the result to the file if valid
            with open(save_result_dir, 'a') as f:
                f.write(f'{res}({acep})\n')
                
    except Exception as e:
        print(f"An error occurred: {e}")  # Print any errors encountered
```
