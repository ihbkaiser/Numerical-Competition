```python
# if you want to run MSHO with function number k (1-24), then you need to set prob = GNBG(k)
algo = MSHO(name="IHBKaiser",prob=GNBG(24), gen_length=30, MAX_FES=500000, BASE_POPSZ=100, learning=True, dynamic_pop=True, phasethree=True)
algo.run()
```

```python
for prob_idx in [18]:
    for seed in range(42, 73):
        np.random.seed(seed)
        random.seed(seed)
        file_name = f'{prob_idx}.dat'
        algo = MSHO(name="IHBKaiser",prob=GNBG(prob_idx), gen_length=30, MAX_FES=500000, BASE_POPSZ=100, learning=True, dynamic_pop=True, phasethree=True)
        res = algo.run()
        with open(file_name, 'a') as f:
                f.write(res + '\n')
```
